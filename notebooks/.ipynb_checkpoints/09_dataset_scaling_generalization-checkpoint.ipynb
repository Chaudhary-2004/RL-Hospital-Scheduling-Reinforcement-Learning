```python
import sys, os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

from collections import deque

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, PROJECT_ROOT)

from utils.hospital_env import HospitalEnv

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
```

```python
def generate_data(n, seed=42):
    np.random.seed(seed)
    DAY = 8 * 60

    data = pd.DataFrame({
        "arrival_time": np.random.randint(0, DAY, n),
        "slot_time": np.random.randint(0, DAY, n),
        "priority": np.random.choice([0,1], n, p=[0.8,0.2]),
        "no_show_prob": np.random.uniform(0.05, 0.4, n),
    })
    data["show"] = (np.random.rand(n) > data["no_show_prob"]).astype(int)
    return data
```

```python
def normalize(state):
    a, s, p, n = state
    return np.array([a/480, s/480, p, n], dtype=np.float32)
```

```python
class DQN(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(4, 64),
            nn.ReLU(),
            nn.Linear(64, 2)
        )

    def forward(self, x):
        return self.net(x)
```

```python
def train_ddqn(data, episodes=50):
    env = HospitalEnv(data)

    policy = DQN()
    target = DQN()
    target.load_state_dict(policy.state_dict())

    optimizer = optim.Adam(policy.parameters(), lr=0.001)
    memory = deque(maxlen=5000)

    gamma = 0.9
    epsilon = 0.2
    batch_size = 64
    target_update = 5

    for ep in range(episodes):
        state = normalize(env.reset()).astype(np.float32)
        done = False

        while not done:
            # -------- Action selection --------
            if random.random() < epsilon:
                action = random.randint(0, 1)
            else:
                with torch.no_grad():
                    q_vals = policy(
                        torch.from_numpy(state).float().unsqueeze(0)
                    )
                    action = torch.argmax(q_vals).item()

            # -------- Environment step --------
            next_state, reward, done = env.step(action)
            next_state = (
                normalize(next_state).astype(np.float32)
                if not done else
                np.zeros(4, dtype=np.float32)
            )

            # -------- Store transition (ONLY NumPy) --------
            memory.append((state, action, reward, next_state, done))
            state = next_state

            # -------- Training step --------
            if len(memory) >= batch_size:
                batch = random.sample(memory, batch_size)
                s, a, r, ns, d = zip(*batch)

                s  = torch.from_numpy(np.stack(s)).float()
                ns = torch.from_numpy(np.stack(ns)).float()
                a  = torch.tensor(a, dtype=torch.long)
                r  = torch.tensor(r, dtype=torch.float32)
                d  = torch.tensor(d, dtype=torch.float32)

                # Q(s,a)
                q_pred = policy(s).gather(1, a.unsqueeze(1)).squeeze()

                # -------- DDQN target --------
                with torch.no_grad():
                    next_actions = policy(ns).argmax(1)
                    q_next = target(ns).gather(
                        1, next_actions.unsqueeze(1)
                    ).squeeze()

                q_target = r + gamma * q_next * (1 - d)

                loss = nn.MSELoss()(q_pred, q_target)

                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

        # -------- Target network update --------
        if ep % target_update == 0:
            target.load_state_dict(policy.state_dict())

    # -------- Final evaluation --------
    return evaluate(policy, HospitalEnv(data))
```

```python
def evaluate(model, env):
    s = normalize(env.reset())
    done = False
    total = 0

    while not done:
        with torch.no_grad():
            a = torch.argmax(model(torch.tensor(s))).item()
        s, r, done = env.step(a)
        if not done:
            s = normalize(s)
        total += r

    return total
```

```python
sizes = [500, 2000, 5000]
results = {}

for size in sizes:
    print(f"\nTraining on dataset size: {size}")
    data = generate_data(size)
    reward = train_ddqn(data)
    results[size] = reward
    print(f"Evaluation Reward: {reward}")
```

```python
plt.plot(results.keys(), results.values(), marker='o')
plt.xlabel("Dataset Size")
plt.ylabel("Total Reward")
plt.title("Phase 9: Dataset Scaling & Generalization")
plt.show()
```
