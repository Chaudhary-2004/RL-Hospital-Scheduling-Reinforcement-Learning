```python
import sys
import os
import pandas as pd

# -----------------------------------
# Project Path Setup (Clean Version)
# -----------------------------------
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

from utils.hospital_env import HospitalEnv

# -----------------------------------
# Load Dataset
# -----------------------------------
DATA_PATH = os.path.join(PROJECT_ROOT, "data", "synthetic_hospital_data.csv")
data = pd.read_csv(DATA_PATH)

# -----------------------------------
# Initialize Environment
# -----------------------------------
env = HospitalEnv(data)

state = env.reset()

print("========== PHASE 2 : ENVIRONMENT TEST ==========")
print("Initial State:", state)
print("State Dimension:", len(state))
print("-------------------------------------------------")

done = False
total_reward = 0
step_count = 0

# -----------------------------------
# Always Accept Policy (Stress Test)
# -----------------------------------
while not done:

    action = 1  # Always accept

    next_state, reward, done = env.step(action)

    print(
        f"Step {step_count:3d} | "
        f"Reward: {reward:4d} | "
        f"ICU Load: {env.current_icu_occupied}/{env.icu_capacity}"
    )

    total_reward += reward
    step_count += 1

print("-------------------------------------------------")
print("Episode finished.")
print("Total reward:", total_reward)
print("Total steps :", step_count)
print("=================================================")
```
