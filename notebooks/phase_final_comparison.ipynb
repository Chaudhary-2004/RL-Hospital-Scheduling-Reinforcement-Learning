```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# ---------------------------------------
# Combine All Algorithm Performance Files
# ---------------------------------------
files = [
    "tabular_q_performance_row.csv",
    "linear_q_performance_row.csv",
    "dqn_performance_row.csv",
    "ddqn_performance_row.csv"
]

df = pd.concat([pd.read_csv(f) for f in files])

# ---------------------------------------
# Calculate Efficiency Metric
# ---------------------------------------
df["Efficiency"] = (
    df["Resource Utilization"] / df["Resource Utilization"].max()
) * 100

# Save final comparison table
df.to_csv("algorithm_comparison.csv", index=False)

print("\nFinal Algorithm Comparison Table:")
print(df)

# Create results folder
os.makedirs("final_results", exist_ok=True)

# ---------------------------------------
# Reward Comparison
# ---------------------------------------
plt.figure()
plt.bar(df["Algorithm"], df["Avg Reward"])
plt.xlabel("Algorithm")
plt.ylabel("Average Reward")
plt.title("Algorithm Reward Comparison")
plt.savefig("final_results/algorithm_reward_comparison.png")
plt.close()

# ---------------------------------------
# Waiting Time Comparison
# ---------------------------------------
plt.figure()
plt.bar(df["Algorithm"], df["Waiting Time"])
plt.xlabel("Algorithm")
plt.ylabel("Average Waiting Time")
plt.title("Waiting Time Comparison")
plt.savefig("final_results/waiting_time_comparison.png")
plt.close()

# ---------------------------------------
# Resource Utilization Comparison
# ---------------------------------------
plt.figure()
plt.bar(df["Algorithm"], df["Resource Utilization"])
plt.xlabel("Algorithm")
plt.ylabel("Resource Utilization (%)")
plt.title("Resource Utilization Comparison")
plt.savefig("final_results/resource_utilization_comparison.png")
plt.close()

# ---------------------------------------
# Convergence Comparison
# ---------------------------------------
plt.figure()
plt.bar(df["Algorithm"], df["Convergence Episode"])
plt.xlabel("Algorithm")
plt.ylabel("Episodes to Converge")
plt.title("Convergence Comparison")
plt.savefig("final_results/convergence_comparison.png")
plt.close()

# ---------------------------------------
# Efficiency Comparison
# ---------------------------------------
plt.figure()
plt.bar(df["Algorithm"], df["Efficiency"])
plt.xlabel("Algorithm")
plt.ylabel("Efficiency (%)")
plt.title("Scheduling Efficiency Comparison")
plt.savefig("final_results/efficiency_comparison.png")
plt.close()

print("\nAll comparison graphs saved in 'final_results' folder.")
```
