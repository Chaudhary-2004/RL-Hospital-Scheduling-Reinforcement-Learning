```python
import pandas as pd
import matplotlib.pyplot as plt

tabular = pd.read_csv("tabular_q_performance_row.csv")
linear  = pd.read_csv("linear_q_performance_row.csv")
dqn     = pd.read_csv("dqn_performance_row.csv")
ddqn    = pd.read_csv("ddqn_performance_row.csv")

df = pd.concat([tabular, linear, dqn, ddqn])

print(df)
```

```python
plt.figure(figsize=(6,4))
plt.bar(df["Algorithm"], df["Waiting Time"])
plt.xlabel("Algorithm")
plt.ylabel("Waiting Time (minutes)")
plt.title("Waiting Time Comparison")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("waiting_time_curve.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(df["Algorithm"], df["Resource Utilization"])
plt.xlabel("Algorithm")
plt.ylabel("Resource Utilization (%)")
plt.title("Resource Utilization Comparison")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("resource_utilization.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(df["Algorithm"], df["Avg Reward"])
plt.xlabel("Algorithm")
plt.ylabel("Average Reward")
plt.title("Algorithm Reward Comparison")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("algorithm_comparison.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(df["Algorithm"], df["Convergence Episode"], marker='o')
plt.xlabel("Algorithm")
plt.ylabel("Convergence Episodes")
plt.title("Convergence Comparison")
plt.grid(True)
plt.tight_layout()
plt.savefig("convergence_comparison.png", dpi=300)
plt.show()
```

```python
log = pd.read_csv("ddqn_training_log.csv")

plt.figure(figsize=(6,4))
plt.plot(log["episode"], log["reward"])
plt.xlabel("Training Episodes")
plt.ylabel("Cumulative Reward")
plt.title("Reward vs Training Episodes")
plt.grid(True)
plt.tight_layout()
plt.savefig("reward_curve.png", dpi=300)
plt.show() 
```
