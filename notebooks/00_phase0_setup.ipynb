```python
import numpy as np
import random
import matplotlib.pyplot as plt

SEED = 42
random.seed(SEED)
np.random.seed(SEED)

print("RL environment setup complete")
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title("Sanity Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```
